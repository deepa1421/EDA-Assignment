# Finding Missing values and converting them to percentage

```python
#creating a function to find percentage of the  missing values that can be used again
def null_values(df): 
    return round((df.isnull().sum()*100/len(df)).sort_values(ascending=False),2)
```

```python
null_values(df)
```

```python
len(null_values(df))
```

# Remove columns with high missing values

```python
#Let us find the columns which has missing values more than 50%. Storing the value in a new variable.
#There are two ways to deal with the missing Values
#If the percentage is very high it's suggested to drop the columns using drop().
#If the percentage is low we can try to fill values with statistical data like mean median mode etc. using fillna()
```

```python
df_missing_values_50 = null_values(df)[null_values(df)>=50]
df_missing_values_50
```

```python
#Number of columns having missing values more than 50%
len(df_missing_values_50)
```

```python
#There are 41 columns with null values more than 50% that needs to be dropped.
```

```python
#Lets get the name of the columns that needs to be dropped.
df_missing_values_50.index
```

```python
df.drop(columns= df_missing_values_50.index, inplace=True)
```

```python
df.shape
```

```python
# we can see that now instead of 122 columns we have only 81 columns
```

```python
df_missing_values_15 = null_values(df)[null_values(df)<=15]
df_missing_values_15.head()
```

```python
# As we can see from the above output 5 columns with null values less than 15%. These are the day,week,month,quarter and year of the AMT_REQ_CREDIT_BUREAU. These columns have no realtion with the Target column so it can be imputed. 
```

```python
 fig = plt.figure(figsize=(18,6))
missing_data = pd.DataFrame((df.isnull().sum())*100/df.shape[0]).reset_index()
missing_data["type"] = "df"
ax = sns.pointplot(x="index",y=0,data=missing_data,hue="type")
plt.xticks(rotation =90,fontsize =7)
plt.title("Percentage of Missing values in Application Data",fontsize=25)
plt.ylabel("PERCENTAGE")
plt.xlabel("COLUMNS")
plt.show()
```
